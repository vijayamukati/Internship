```python
#importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings("ignore")

driver=webdriver.Chrome(r"C:\Drivers\webdriver//chromedriver.exe")

url='https://www.olx.in/cars_c84'
driver.get(url)

#Creating empty lists
brand=[]
year=[]
fuel=[]
transmission=[]
kms=[]
price=[]
prod_URL=[]
location=[]

#Getting the URL of the cars
for i in range(0,500):

    url=driver.find_elements_by_xpath('//img[@class="_1N079"]')
    for i in url:
        prod_URL.append(i.get_attribute('src'))
    try:
        next_btn=driver.find_element_by_xpath('//div[@class="JbJAl"]//span').click()
    except NoSuchElementException:
        pass    
        

for i in prod_URL:
    driver.get(i)
    
    try:
        brand.append(driver.find_elements_by_xpath('//div[@class="_4aNdc"]').text)                                                                                                                                                                                                  
    except NoSuchElementException as e:
        brand.append("-")
    try: 
        price.append(driver.find_elements_by_xpath('//span[@class="_3GOwr"][1]').text)  
    except NoSuchElementException as e:
        price.append("-")
    
    try:
        year.append(driver.find_elements_by_xpath('//div[@class="KFHpP"]').text)
    except NoSuchElementException as e:
        year.append("-")
    try:
        location.append(driver.find_elements_by_xpath('//div[@class="_1zvfB"]').text) 
    except NoSuchElementException as e:
        location.append("-")   
    
    try:
        kms.append(driver.find_elements_by_xpath('//div[@class="_1u27F"][2]').text)
    except NoSuchElementException as e:
        kms.append("-")
    try:         
        fuel.append(driver.find_elements_by_xpath('//div[@class="_1u27F"][1]').text)
    except NoSuchElementException as e:
        fuel.append("-")
    try:    
        transmission.append(driver.find_elements_by_xpath('//div[@class="_1u27F"][3]').text)
    except NoSuchElementException as e:
        transmission.append("-")  

#Creating a dataframe and checking the data extracted
olx=pd.DataFrame({'Brand':brand[0:5000],'Year':year[0:5000],'Price':price[0:5000],'Km_driven':kms[0:5000],'Fuel':fuel[0:5000],'Transmission':transmission[0:5000],'LOCATION with DATE':location[0:5000]})

olx

olx.to_csv('olxdata.csv', index=False, encoding='utf-8')
```
