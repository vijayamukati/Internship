```python
#importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Chrome(r"C:\Drivers\webdriver//chromedriver.exe")
```

```python
url='https://www.olx.in/cars_c84'
driver.get(url)
```

```python
brand=[]
price=[]
year=[]
location=[]
```

```python
for i in range(0,50):
    
    brand_tags=driver.find_elements_by_xpath('//div[@class="_4aNdc"]')                                                                                                                                                                                                  
    for i in brand_tags:
        brand.append(i.text)
     
    p=driver.find_elements_by_xpath('//span[@class="_3GOwr"][1]')  
    for i in p:
        price.append(i.text)

    y=driver.find_elements_by_xpath('//div[@class="KFHpP"]')
    for i in y:
        year.append(i.text)
    
    loc=driver.find_elements_by_xpath('//div[@class="_1zvfB"]')
    for i in loc:
        location.append(i.text) 
        
   
```

```python
#Creating a dataframe and checking the data extracted
olx=pd.DataFrame({'NAME':brand[0:5000],'YEAR WITH TOTAL KMS':year[0:5000],'SELLING PRICE':price[0:5000],'LOCATION with DATE':location[0:5000]})
```
